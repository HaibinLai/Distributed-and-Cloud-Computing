```python
import matplotlib.pyplot as plt

# Processes and sizes
procs = [1, 2, 4, 8, 16, 32]

data = {
    500: [0.467676, 0.238192, 0.119467, 0.061727, 0.033172, 0.021730],
    250: [0.059059, 0.032358, 0.015812, 0.008433, 0.004888, 0.008457],
    100: [0.003769, 0.004256, 0.001356, 0.000875, 0.000771, 0.006969],
    50:  [0.000504, 0.000545, 0.000416, 0.000414, 0.001147, 0.006003],
    10:  [0.000025, 0.000247, 0.000284, 0.000253, 0.000458, 0.004371],
}

# 字体15
plt.rcParams.update({'font.size': 16})

# Compute speedup for each matrix size
speedup_data = {size: [times[0] / t for t in times] for size, times in data.items()}

plt.figure(figsize=(11,10))

# Subplot 1: Execution time
plt.subplot(2,1,1)
for size, times in sorted(data.items(), reverse=True):
    plt.plot(procs, times, marker='o', linewidth=2, markersize=6, label=f"N={size}")
plt.xlabel("Number of Processes")
plt.ylabel("Time (s)")
plt.title("MPI Matmul Scaling", fontsize=20, fontweight="bold")
plt.grid(linestyle="--", alpha=0.5)
plt.legend(loc="upper right", ncol=1, title="Matrix Size")
plt.xticks(procs)

# Subplot 2: Speedup
plt.subplot(2,1,2)
for size, speedups in sorted(speedup_data.items(), reverse=True):
    plt.plot(procs, speedups, marker='s', linewidth=2, markersize=6, label=f"N={size}")
plt.plot(procs, procs, linestyle="--", color="gray", label="Ideal Linear Speedup")
plt.xlabel("Number of Processes")
plt.ylabel("Speedup")
plt.grid(linestyle="--", alpha=0.5)
plt.legend(loc="upper left", ncol=2)
plt.xticks(procs)

plt.tight_layout()
# plt.savefig("/mnt/data/mpi_matmul_time_speedup.png", dpi=220, bbox_inches="tight")
# plt.savefig("/mnt/data/mpi_matmul_time_speedup.pdf", bbox_inches="tight")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Processes and sizes
procs = [1, 2, 4, 8, 16, 32]

# MPI data
data = {
    500: [0.467676, 0.238192, 0.119467, 0.061727, 0.033172, 0.021730],
    250: [0.059059, 0.032358, 0.015812, 0.008433, 0.004888, 0.008457],
    100: [0.003769, 0.004256, 0.001356, 0.000875, 0.000771, 0.006969],
    50:  [0.000504, 0.000545, 0.000416, 0.000414, 0.001147, 0.006003],
    10:  [0.000025, 0.000247, 0.000284, 0.000253, 0.000458, 0.004371],
}

# Baseline brute-force single-thread times
baseline = {
    10: 0.000004,
    50: 0.000580,
    100: 0.004633,
    250: 0.057622,
    500: 0.455162,
}

plt.rcParams.update({'font.size': 16})

# Compute MPI speedup (vs MPI 1-process)
mpi_speedup = {size: [times[0] / t for t in times] for size, times in data.items()}

# Compute speedup vs baseline brute-force
baseline_speedup = {
    size: [baseline[size] / t for t in times] for size, times in data.items()
}

# plt.figure(figsize=(11,13))

# 画布，左右两个子图
fig, axes = plt.subplots(1, 2, figsize=(20,9))

# --- 左图: Execution time ---
ax = axes[0]
for size, times in sorted(data.items(), reverse=True):
    ax.plot(procs, times, marker='o', linewidth=2, markersize=6, label=f"MPI N={size}")
# baseline 画成红色虚线
for size, t in baseline.items():
    ax.hlines(t, procs[0], procs[-1], linestyles="--", colors="red", linewidth=2, label=f"Baseline N={size}")
ax.set_xlabel("Number of Processes")
ax.set_ylabel("Time (s)")
ax.set_title("Execution Time", fontsize=18, fontweight="bold")
ax.grid(linestyle="--", alpha=0.5)
ax.legend(loc="upper right", ncol=2, fontsize=16, title="Matrix Size / Method")
# log y
ax.set_yscale("log")
ax.set_xticks(procs)

# --- 右图: Speedup ---
ax = axes[1]
for size, speedups in sorted(mpi_speedup.items(), reverse=True):
    ax.plot(procs, speedups, marker='s', linewidth=2, markersize=6, label=f"MPI N={size}")
# for size, speedups in sorted(baseline_speedup.items(), reverse=True):
#     ax.plot(procs, speedups, marker='^', linewidth=2, markersize=6, linestyle=":", label=f"Baseline N={size}")
ax.plot(procs, procs, linestyle="--", color="gray", label="Ideal Linear Speedup")
ax.set_xlabel("Number of Processes")
ax.set_ylabel("Speedup compared to MPI 1-process")
ax.set_title("Speedup", fontsize=18, fontweight="bold")
ax.grid(linestyle="--", alpha=0.5)
ax.legend(loc="upper left", ncol=2, fontsize=16)
ax.set_xticks(procs)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import LogNorm

# Processes and sizes
procs = [1, 2, 4, 8, 16, 32]
sizes = [500, 250, 100, 50, 10]

data = {
    500: [0.467676, 0.238192, 0.119467, 0.061727, 0.033172, 0.021730],
    250: [0.059059, 0.032358, 0.015812, 0.008433, 0.004888, 0.008457],
    100: [0.003769, 0.004256, 0.001356, 0.000875, 0.000771, 0.006969],
    50:  [0.000504, 0.000545, 0.000416, 0.000414, 0.001147, 0.006003],
    10:  [0.000025, 0.000247, 0.000284, 0.000253, 0.000458, 0.004371],
}

# Convert to matrix
Z = np.array([data[size] for size in sizes])

plt.figure(figsize=(9,6))
im = plt.imshow(Z, aspect="auto", origin="lower", 
                extent=[min(procs), max(procs), min(sizes), max(sizes)],
                cmap="viridis", norm=LogNorm())

plt.colorbar(im, label="Time (s, log scale)")
plt.xticks(procs)
plt.yticks(sizes)
plt.xlabel("Number of Processes")
plt.ylabel("Matrix Size (N)")
plt.title("MPI Matmul Execution Time Heatmap (Log Scale)", fontsize=14, fontweight="bold")

plt.tight_layout()
# plt.savefig("/mnt/data/mpi_matmul_heatmap_log.png", dpi=220, bbox_inches="tight")
# plt.savefig("/mnt/data/mpi_matmul_heatmap_log.pdf", bbox_inches="tight")
plt.show()
```
